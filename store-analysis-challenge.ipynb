### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

lojas = pd.concat([loja1, loja2, loja3, loja4])
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

lojas.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja1["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()
print(f"Faturamento Loja 1: R$ {faturamento_loja1:.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:.2f}")
```

```python
faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4
print(f"Faturamento Total das 4 lojas: R$ {faturamento_total:.2f}")
```

# 2. Vendas por Categoria


```python
vendas_categoria_loja1 = loja1["Categoria do Produto"].value_counts()
vendas_categoria_loja2 = loja2["Categoria do Produto"].value_counts()
vendas_categoria_loja3 = loja3["Categoria do Produto"].value_counts()
vendas_categoria_loja4 = loja4["Categoria do Produto"].value_counts()
print("Vendas por Categoria - Loja 1:")
print(vendas_categoria_loja1)
print("\nVendas por Categoria - Loja 2:")
print(vendas_categoria_loja2)
print("\nVendas por Categoria - Loja 3:")
print(vendas_categoria_loja3)
print("\nVendas por Categoria - Loja 4:")
print(vendas_categoria_loja4)
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja1["Avaliação da compra"].mean()
media_avaliacao_loja2 = loja2["Avaliação da compra"].mean()
media_avaliacao_loja3 = loja3["Avaliação da compra"].mean()
media_avaliacao_loja4 = loja4["Avaliação da compra"].mean()
print(f"Média de Avaliação - Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação - Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação - Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação - Loja 4: {media_avaliacao_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
def exibir_destaques_vendas(loja, nome_loja):
    vendas = loja["Produto"].value_counts()
    mais_vendido = vendas.idxmax()
    menos_vendido = vendas.idxmin()
    print(f"\n📊 {nome_loja}")
    print("Produto mais vendido:", mais_vendido, "-", vendas.max(), "vendas")
    print("Produto menos vendido:", menos_vendido, "-", vendas.min(), "vendas")
```

```python
exibir_destaques_vendas(loja1, "Loja 1")
exibir_destaques_vendas(loja2, "Loja 2")
exibir_destaques_vendas(loja3, "Loja 3")
exibir_destaques_vendas(loja4, "Loja 4")
```

#Gerando gráfico

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
faturamentos = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

lojas = list(faturamentos.keys())
valores = list(faturamentos.values())
cores = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA']

fig, ax = plt.subplots()

bars = ax.bar(lojas, valores, color=cores)

for bar, valor in zip(bars, valores):
    altura = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, altura,
            f'R$ {valor:,.2f}', ha='center', va='bottom')

ax.set_title('Faturamento Total por Loja')
ax.set_ylabel('Faturamento (R$)')
ax.set_xlabel('Lojas')

plt.show()
```

```python
lojas = [loja1, loja2, loja3, loja4]
nomes = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

for i, loja in enumerate(lojas):
    categorias = loja['Categoria do Produto'].value_counts()

    fig, ax = plt.subplots()
    # Criar gráfico de pizza com "hole" para efeito donut
    wedges, texts, autotexts = ax.pie(
        categorias.values,
        labels=categorias.index,
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.85,
        wedgeprops=dict(width=0.8, edgecolor='black')
    )
    centre_circle = plt.Circle((0, 0), 0.50, fc='white')
    fig.gca().add_artist(centre_circle)

    ax.set_title(f"Distribuição de Categorias - {nomes[i]}", pad=20)
    plt.text(0, 0, nomes[i], horizontalalignment='center', verticalalignment='center', fontsize=14, fontweight='bold')

    plt.axis('equal')
    plt.show()
```

```python
fretes = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

lojas = list(fretes.keys())
valores = list(fretes.values())

fig, ax = plt.subplots()

ax.plot(lojas, valores, marker='o', color='darkblue', linewidth=3, markersize=10, markerfacecolor='orange')

ax.set_title('Custo Médio de Frete por Loja')
ax.set_xlabel('Lojas')
ax.set_ylabel('Frete Médio (R$)')

ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

#Análise Geográfica das Vendas

##Objetivo
Explorar as coordenadas geográficas (latitude e longitude) das vendas para identificar padrões regionais que possam influenciar o desempenho de cada loja. A análise busca entender se:

* Há regiões com maior concentração de vendas.

* Algumas lojas se destacam em regiões específicas.

* Existe correlação entre localização e faturamento/avaliações.




###Mapa de Dispersão por Loja (Scatter Geo)

```python
df = pd.concat([
    loja1.assign(Loja='Loja 1'),
    loja2.assign(Loja='Loja 2'),
    loja3.assign(Loja='Loja 3'),
    loja4.assign(Loja='Loja 4')
], ignore_index=True)

plt.figure(figsize=(12, 8))

for loja in df["Loja"].unique():
    dados = df[df["Loja"] == loja]
    plt.scatter(dados["lon"], dados["lat"], label=loja, alpha=0.6)

plt.title("Mapa de Dispersão por Loja")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.grid(True)
plt.show()
```

###Análise Geográfica do Faturamento e Avaliações

```python
df["Faturamento"] = df["Preço"] + df["Frete"]

agrupado = df.groupby(["lat", "lon"], as_index=False).agg({
    "Faturamento": "sum",
    "Avaliação da compra": "mean"
})

plt.figure(figsize=(12, 8))
sc = plt.scatter(agrupado["lon"], agrupado["lat"],
                 s=agrupado["Faturamento"] / 20,
                 c=agrupado["Avaliação da compra"], cmap='viridis', alpha=0.7)

plt.colorbar(sc, label="Avaliação Média")
plt.title("Faturamento e Avaliação por Localização")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
